```python
from matplotlib import pylab
import numpy as np
x=np.linspace(0,10,25)
x
```

```python
y=x*x+2
y
```

```python
pylab.plot(x,y, 'b')
```

```python
pylab.subplot(1,2,1)
pylab.plot(x,y, 'r--')
pylab.subplot(2,2,2)
pylab.plot(y,x,'g-*')
```

```python
from matplotlib import pyplot as plt
fig=plt.figure()
axis=fig.add_axes([0.5,0.1,2,2])
axis.plot(x,y,'g')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=3)
for ax in axes:
    ax.plot(x,y,'r')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'r')
axes2.plot(x,y,'g')
```

```python
fig,axes=plt.subplots(dpi=150)
axes.plot(x,x+1,color='blue')
axes.plot(x,x**2,color='green',alpha=.5)
axes.plot(x,x+5,color='blue',linewidth=0.25)
axes.plot(x,x+2,color='red',lw=2,linestyle='-')
axes.plot(x,x+3,color='green',marker='o',markerfacecolor='red',markersize='8')
```

```python
fig,axes=plt.subplots(1,2,figsize=(10,5))
axes[0].plot(x,x**2,x,x**3,lw=2)
axes[0].grid(True)

axes[1].plot(x,x**2,x,x**3,lw=2)
axes[1].set_ylim([0,60])
axes[1].set_xlim([2,5])
```

```python
n=np.array([0,1,2,3,4,5])

fig,axes=plt.subplots(1,4,figsize=(16,5))
axes[0].scatter(x,x+0.25*np.random.randn (len(x)))
axes[1].step(n,n**2,lw=2)
axes[2].bar(n,n**2,align="center",width=0.5,alpha=0.5)
axes[3].fill_between(x,x**2,x**3,color="green",alpha=0.5)
```
